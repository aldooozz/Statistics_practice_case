```python
# import library
import pandas as pd
import numpy as np
```

```python
#import data
data = pd.read_csv('Seasons_Stats.csv')
data
```

```python
data.duplicated().any()
```

```python
#subset year 2017
data_2017 = data[data["Year"] == 2017]
data_2017
```

```python
# drop duplicate
data_2017 = data_2017.drop_duplicates(subset=['Player'],keep='last').reset_index()
data_2017
```

```python
data_2017.info()
```

```python
data_2017.isnull().sum()
```

```python
data_2017 = data_2017.drop(['blanl', 'blank2','No','index'], axis = 1)
data_2017
```

```python
data_2017.describe()
```

```python
data_2017_Age = data_2017[['Player','Age','Tm']]
data_2017_Age_Young = data_2017_Age.groupby(['Tm']).min().reset_index()
data_2017_Age_Old = data_2017_Age.groupby(['Tm']).max().reset_index()
```

```python
data_2017_Age_Young
```

```python
data_2017_tm= data_2017[['Tm','TRB%','AST%','BLK%']]
data_2017_tm
```

```python
data_2017_tm = data_2017_tm.groupby(['Tm']).mean().reset_index()
data_2017_tm
```

```python
data_2017_tm_TRB = data_2017_tm[['Tm','TRB%']].sort_values(by='TRB%',ascending=False).head(1)
data_2017_tm_TRB
```

```python
data_2017_tm_AST = data_2017_tm[['Tm','AST%']].sort_values(by='AST%',ascending=False).head(1)
data_2017_tm_AST
```

```python
data_2017_tm_BLK = data_2017_tm[['Tm','BLK%']].sort_values(by='BLK%',ascending=False).head(1)
data_2017_tm_BLK
```
